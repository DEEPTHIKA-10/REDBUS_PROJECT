```python
# Load the dataset for Andhra Pradesh
df_a = pd.read_csv(r"C:\\Users\\kumar\\Desktop\\scraped_data\\allcombineddata\\df_a.csv")

# Initialize WebDriver
chrome_options = Options()
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("start-maximized")
driver_A = webdriver.Chrome(options=chrome_options)

# Lists for storing data
Bus_names_A = []
Bus_types_A = []
Start_Time_A = []
End_Time_A = []
Ratings_A = []
Total_Duration_A = []
Prices_A = []
Seats_Available_A = []
Route_names_A = []
Route_links_A = []

# Loop through each route in the dataset
for i, r in df_a.iterrows():
    link = r["routelink"]
    route = r["route"]

    driver_A.get(link)
    time.sleep(5)  # Wait for page to load

    # Click on elements to reveal bus details
    try:
        elements = driver_A.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
        for element in elements:
            element.click()
            time.sleep(3)
    except:
        pass  

    # Click to view bus details
    try:
        clicks = driver_A.find_element(By.XPATH, "//div[@class='button']")
        clicks.click()
        time.sleep(3)
    except:
        pass  

    # Scroll page to load all buses
    scrolling = True
    while scrolling:
        old_page_source = driver_A.page_source
        ActionChains(driver_A).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(3)
        new_page_source = driver_A.page_source
        if new_page_source == old_page_source:
            scrolling = False

    # Extract bus details
    bus_name = [b.text for b in driver_A.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")]
    bus_type = [b.text for b in driver_A.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")]
    start_time = [b.text for b in driver_A.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")]
    end_time = [b.text for b in driver_A.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")]
    total_duration = [b.text for b in driver_A.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")]
    price = [b.text for b in driver_A.find_elements(By.XPATH, '//*[@class="fare d-block"]')]
    seats = [b.text for b in driver_A.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")]

    # Extract Ratings
    try:
        wait = WebDriverWait(driver_A, 10)
        ratings_elements = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[contains(@class,'column-six') and contains(@class,'p-right-10')]") ))
        ratings = [b.text.strip() if b.text.strip() else "N/A" for b in ratings_elements]
    except:
        ratings = ["N/A"] * len(bus_name)

    # Ensure all lists have the same length
    for b_name, b_type, s_time, e_time, duration, rate, p, s in zip_longest(bus_name, bus_type, start_time, end_time, total_duration, ratings, price, seats, fillvalue="N/A"):
        Bus_names_A.append(b_name)
        Bus_types_A.append(b_type)
        Start_Time_A.append(s_time)
        End_Time_A.append(e_time)
        Total_Duration_A.append(duration)
        Ratings_A.append(rate)
        Prices_A.append(p)
        Seats_Available_A.append(s)
        Route_links_A.append(link)
        Route_names_A.append(route)

# Creating DataFrame
data_A = {
    'Bus_name': Bus_names_A,
    'Bus_type': Bus_types_A,
    'Start_time': Start_Time_A,
    'End_time': End_Time_A,
    'Total_duration': Total_Duration_A,
    'Price': Prices_A,
    'Seats_Available': Seats_Available_A,
    'Ratings': Ratings_A,
    'Route_link': Route_links_A,
    'Route_name': Route_names_A
}

df_andhra = pd.DataFrame(data_A)

# Save the DataFrame to CSV
output_path = r"C:/Users/kumar/Desktop/scraped_data/andhra_busdetails.csv"
df_andhra.to_csv(output_path, index=False)

print("Scraping complete! Data saved successfully.")
print(df_andhra)
```
